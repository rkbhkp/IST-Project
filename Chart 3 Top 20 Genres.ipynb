```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dat = pd.read_pickle("pickle1/tags.pkl")

dat.head(5)
```

```python
N = 20
df = pd.Series(np.random.randint(10,50,N), index=np.arange(1,N+1))
cmap = plt.cm.tab10
colors = cmap(np.arange(len(df)) % cmap.N)

a = dat['tag'].value_counts()[:20].plot(kind='barh', 
                                        color=colors, 
                                        alpha=0.9, 
                                        figsize=(8,6), 
                                        legend = False)
plt.xlabel("Number of Occurences")
plt.ylabel("Genre")
plt.title("Top 20 Genres by Number of Reviews")
plt.grid(True, color = "black", alpha=0.2)
plt.gca().invert_yaxis()
```
