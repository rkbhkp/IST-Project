```python
dat_Names = list(dat2['movieId'])

dict_lookup = dict(zip(dat2['movieId'], dat2['title']))

dict_lookup
```

```python
dat1['movieId'] = [dict_lookup[item] for item in dat_Names]
```

```python
N = 20
df = pd.Series(np.random.randint(10,50,N), index=np.arange(1,N+1))
cmap = plt.cm.tab10
colors = cmap(np.arange(len(df)) % cmap.N)

a = dat1['movieId'].value_counts()[:20].plot(kind='barh', 
                                        color=colors, 
                                        alpha=0.9, 
                                        figsize=(8,6), 
                                        legend = False)
plt.xlabel("Number of Reviews")
plt.ylabel("Movie")
plt.title("Top 20 Genres by Number of Reviews")
plt.grid(True, color = "black", alpha=0.2)
plt.gca().invert_yaxis()
```
